```python
#### required packages####
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
import csv

save_fig_path = 'images/'

### read file
file_path = "training.data"
num_data = 0
file = open(file_path, 'r')
input_data = file.readlines()
print ('Number of training data are:', len(input_data))

class1 = []
class2 = []
questions = []
for rows in input_data:
    class_question = rows.split(' ')
    questions.append(class_question[-1])
    class1.append(class_question[0].split(':')[0])
    class2.append(class_question[0].split(':')[1])
print (class1[-1], class2[-1])
```

```python
## visualize data distribution
print (set(class1))
LOC_sub, NUM_sub, ABBR_sub, DESC_sub, ENTY_sub, HUM_sub = {}, {}, {}, {}, {}, {}

def map_class2(class1_dict, class2):
    if class2 in class1_dict:
        class1_dict[class2] += 1
    else:
        class1_dict[class2] = 1

def pie_plot(sub_class, title):
    labels = sub_class.keys()
    sizes = sub_class.values()
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes, labels = labels)
    ax1.set_title(title)
    plt.savefig(save_fig_path + '/' + title + '.jpg')
    
for i in range(len(input_data)):
    if class1[i] == 'LOC':
        map_class2(LOC_sub, class2[i])
    elif class1[i] == 'NUM':
        map_class2(NUM_sub, class2[i])
    elif class1[i] == 'ABBR':
        map_class2(ABBR_sub, class2[i])
    elif class1[i] == 'DESC':
        map_class2(DESC_sub, class2[i])
    elif class1[i] == 'ENTY':
        map_class2(ENTY_sub, class2[i])
    else:
        map_class2(HUM_sub, class2[i])

print ('Number of class 2 in each class 1:')
print ('LOC:', len(LOC_sub), 'NUM:', len(NUM_sub), 'ABBR:', len(ABBR_sub), 'DESC:', len(DESC_sub), 'ENTY:', len(ENTY_sub), \
       'HUM:', len(HUM_sub))
pie_plot(LOC_sub,'LOC')
pie_plot(NUM_sub,'NUM')
pie_plot(ABBR_sub,'ABBR')
pie_plot(DESC_sub,'DESC')
pie_plot(ENTY_sub,'ENTY')
pie_plot(HUM_sub,'HUM')
```
